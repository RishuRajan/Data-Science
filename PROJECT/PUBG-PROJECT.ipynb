```python
#TASK 8
#Comment on distribution of the match's duration. Use seaborn.
#CODE-
sns.set_style('whitegrid')
sns.distplot( dataframe['matchDuration'],kde = True, color ='green');
```

Here, Duration of match is High between 1250 to 1500

```python
#TASK 9
#Comment on distribution of the walk distance. Use seaborn.
#CODE-
sns.set_style("whitegrid")
sns.distplot( dataframe['walkDistance'], kde = True ,color = 'orange')
```

```python
#TASK 10
#Plot distribution of the match's duration vs walk distance one below the other.
#CODE-
%matplotlib inline
plt.style.use('dark_background')
plt.figure()

#ploting for matchDuration
plt.subplot(2,1,1)
plt.plot(dataframe['matchDuration'],'-')
plt.xlabel('matchDuration')

#ploting for walkDistance
plt.subplot(2,1,2)
plt.plot(dataframe['walkDistance'],'--')
plt.xlabel('walkDistance')
```

```python
#TASK 11
#Plot distribution of the match's duration vs walk distance side by side.
#CODE-
%matplotlib inline
plt.style.use('dark_background')
plt.figure()

#ploting for matchDuration
plt.subplot(1,2,1)
plt.plot(dataframe['matchDuration'])
plt.xlabel('matchDuration')

#ploting for walkDistance
plt.subplot(1,2,2)
plt.plot(dataframe['walkDistance'])
plt.xlabel('walkDistance')
```
